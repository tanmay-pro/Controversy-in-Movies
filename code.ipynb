```python
import os
import numpy as np
import pandas as pd
```

```python
comb = pd.read_csv('combined.csv')
```

```python
comb.head()
```

```python
# print all movies with 30% of ratings >4.5 and 30 % of ratings <1.0

polarizedMovies = []

for index, row in comb.iterrows():
    totalMovieRatings  = row['0.5'] + row['1.0'] + row['1.5'] + row['2.0'] + row['2.5'] + row['3.0'] + row['3.5'] + row['4.0'] + row['4.5'] + row['5.0']
    highRatings = row['4.5'] + row['5.0']
    lowRatings = row['0.5'] + row['1.0']
    try:
        if (highRatings/totalMovieRatings >= 0.40) and (lowRatings/totalMovieRatings >= 0.40):
            polarizedMovies.append(row['title'])
    except:
        continue
```
